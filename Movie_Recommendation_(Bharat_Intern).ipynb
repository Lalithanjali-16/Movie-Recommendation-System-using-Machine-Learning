```python
import numpy as np
import pandas as pd
import difflib
#we take movie name as input from the users and there are chances of them giving wrong spellings
#so we need to find the similar match for it so we use difflib
from sklearn.feature_extraction.text import TfidfVectorizer
#conversion of textual data into numerical data , also known as feature vector
from sklearn.metrics.pairwise import cosine_similarity
#to find similar movies based on string(or vector) in movie name
```

Data Collection and Pre-Processing

```python
data=pd.read_csv("TeluguMovies_dataset.csv")
```

```python
selected_features=["Movie","Genre","Rating"]
print(selected_features)
```

```python
for i in selected_features:
  data[i]=data[i].fillna("")
```

```python
combined_features=data['Movie'] + " "+data['Genre'] + " " + data['Rating'].astype(str)
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
#textual data to numerical data
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=input("Enter your favourite movie name : ")
```

```python
list_of_all_titles=data['Movie'].tolist()
print(list_of_all_titles)
#making list of all movie names
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
#all close names based on string in name name
```

```python
close_match=find_close_match[0]
print(close_match)
#the first name among all the closed matches
```

```python
index_of_the_movie=data[data.Movie==close_match]['Index'].values[0]
print(index_of_the_movie)
#we find the index of the closed match name to get the whole row
```

```python
#list(enumerate) is used to find the similarity of one row to all the rows
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#finding movie name based on the index
print("\n\nMovies recommended for you\n\n")
j=1
for i in sorted_similar_movies:
  ind=i[0]
  suggested_movie=data[data.index==ind]["Movie"].values[0]
  if j<=10:
    print(j,suggested_movie)
  j=j+1
```
